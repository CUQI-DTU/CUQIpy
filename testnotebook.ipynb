```python
import numpy as np
import matplotlib.pyplot as plt
from cuqi.testproblem import Deconvolution1D
from cuqi.distribution import Gaussian, Gamma, JointDistribution, GMRF, LMRF, ImplicitRegularizedGaussian,ImplicitRegularizedGMRF
from cuqi.sampler import Gibbs, LinearRTO, Conjugate, UGLA, ConjugateApprox
from cuqi.problem import BayesianProblem

n = 128
A, y_obs, info = Deconvolution1D.get_components(phantom='square', dim=n)

x = ImplicitRegularizedGMRF(np.zeros(n), 200, constraint="box", lower_bound = 0.1, upper_bound = 0.9)
y = Gaussian(A, 0.01**2)

BP = BayesianProblem(y, x)
BP.set_data(y=y_obs)

samples = BP.sample_posterior(500)
samples.plot_ci(exact=info.exactSolution)
```
